## Importação das bibliotecas



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

## Gerando o dataset

```python
np.random.seed(0)
n_samples = 300

data = {
    'valor_medio_gasto': np.random.randn(n_samples) * 20 + 200,
    'frequencia_compras': np.random.randn(n_samples) * 5 + 15,
    'categorias_produtos': np.random.randn(n_samples) * 2 + 8
}

df = pd.DataFrame(data)
```

## Padronizando os dados

```python
scaler = StandardScaler()
data_scaled = scaler.fit_transform(df)
```

## Aplicando o algoritmo de clustering K-means

```python
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(data_scaled)

df['cluster'] = kmeans.labels_
```

## Visualização dos clusters

```python
plt.figure(figsize=(10, 6))
for cluster in range(3):
    cluster_data = df[df['cluster'] == cluster]
    plt.scatter(cluster_data['valor_medio_gasto'], cluster_data['frequencia_compras'], label=f'Cluster {cluster}')

plt.title('Clusters de Clientes')
plt.xlabel('Valor Médio Gasto')
plt.ylabel('Frequência de Compras')
plt.legend()
plt.show()
```

## Justificativa

Foi escolhido este algoritmo K-means para este problema por sua simplicidade e eficiência em dividir os dados em grupos com base em características semelhantes.
Como o objetivo é segmentar clientes com base em comportamentos específicos (valor gasto, frequência de compras e variedade de categorias de produtos), o K-means é adequado, pois funciona bem em dados com distribuição relativamente uniforme e permite criar grupos interpretáveis.

Vantagens:
- É rápido e escalável, ideal para datasets maiores.
- Os clusters gerados podem ser usados diretamente para criar campanhas de marketing segmentadas.

Limitações:
- O K-means não é robusto a outliers, o que pode afetar a qualidade dos clusters.
- Requer a especificação do número de clusters (k) previamente.

Os clusters formados podem revelar grupos como "clientes que gastam muito e compram frequentemente", "clientes ocasionais", ou "clientes com preferências diversificadas". Esses insights ajudam a criar estratégias personalizadas de promoção e fidelização.
